```python
import pandas as pd
```

```python
datafile="/content/WineQT.csv"
data=pd.read_csv(datafile)
data.head()
```

```python
data=data.drop(['Id'], axis=1)
data.head()
```

```python
mean=pd.DataFrame(data=data.mean())
mean=pd.DataFrame.transpose(mean.rename(columns={0: "mean"}))
```

```python
median=pd.DataFrame(data=data.median())
median=pd.DataFrame.transpose(median.rename(columns={0: "median"}))
```

```python
min=pd.DataFrame(data=data.min())
min=pd.DataFrame.transpose(min.rename(columns={0: "min"}))
```

```python
max=pd.DataFrame(data=data.max())
max=pd.DataFrame.transpose(max.rename(columns={0: "max"}))
```

```python
variance=pd.DataFrame(data=data.var())
variance=pd.DataFrame.transpose(variance.rename(columns={0: "variance"}))
```

```python
standardDeviation=pd.DataFrame(data=data.std())
standardDeviation=pd.DataFrame.transpose(standardDeviation.rename(columns={0: "standard deviation"}))
```

```python
skewness=pd.DataFrame(data=data.skew())
skewness=pd.DataFrame.transpose(skewness.rename(columns={0: "skewness"}))
```

```python
kurtosis=pd.DataFrame(data=data.kurtosis())
kurtosis=pd.DataFrame.transpose(kurtosis.rename(columns={0: "kurtosis"}))
```

```python
result = pd.concat([mean, median, min, max, variance, standardDeviation, skewness, kurtosis])
result
```
